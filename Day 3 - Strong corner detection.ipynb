```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("../opencv-np-workshop/data/img/regensburg.jpg")
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
corners = cv2.goodFeaturesToTrack(gray, maxCorners=50, qualityLevel=0.05, minDistance=10)
```

```python
corners.shape
```

```python
for c in corners:
    x, y = c[0] # Center of circle on top of the corner
    cv2.circle(img, (x,y), 10, (0,0,255) , -1)
plt.figure(figsize=(10,10))
plt.imshow(img[:,:,::-1])
```
