```python
text = cv2.imread("../opencv-np-workshop/data/img/poem.PNG", 0)
plt.imshow(text, cmap="gray")
```

```python
_, thr = cv2.threshold(text, 200, 255, cv2.THRESH_BINARY)
plt.imshow(thr, cmap="gray")
```

## Replace global threshold with local

```python
thr_adapt = cv2.adaptiveThreshold(
    src = text, 
    maxValue = 255, 
    adaptiveMethod = cv2.ADAPTIVE_THRESH_MEAN_C, # Calculate mean on neighbourhood of pixel
    thresholdType = cv2.THRESH_BINARY, # Apply binary threshold with above value
    blockSize = 13, # pixel neigbourhood size, should be odd (3,5,7,...)
    C = 10 # Value substracted to everything on the beginning
)
```

```python
plt.imshow(thr_adapt, cmap="gray")
```
