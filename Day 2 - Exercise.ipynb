```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("./data/img/regensburg.jpg") 
# TIP: adding 0 as argument reads the image as grayscale
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
_, img_thr = cv2.threshold(gray,100,255, cv2.THRESH_BINARY)
# if > 100 then 255 else 0
plt.subplot(1,2,1)
plt.imshow(gray, cmap="gray")
plt.subplot(1,2,2)
plt.imshow(img_thr, cmap="gray")
```

```python
img_thr.max()
```

```python
img_thr.min()
```
