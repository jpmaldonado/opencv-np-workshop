```python
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
nemo = cv2.imread("../opencv-np-workshop/data/img/nemo.jpg")
world = cv2.imread("../opencv-np-workshop/data/img/nemo-sea.jpg")
```

# Strategy

- Identify keypoints and descriptor vectors in both Nemo and the world.
- Match keypoints between Nemo and the world.
- Visualize those matches.

```python
detector = cv2.ORB_create(nfeatures=30) #Default = 500
```

```python
kp_nemo, des_nemo = detector.detectAndCompute(nemo, mask=None)
kp_world, des_world = detector.detectAndCompute(world, mask=None)
```

Once we have keypoints and descriptors, we need to match.

```python
matcher = cv2.BFMatcher_create(cv2.NORM_HAMMING2)
matches = matcher.match(des_nemo,des_world)
```

```python
out = cv2.drawMatches(
    nemo, kp_nemo, 
    world, kp_world,
    matches, None #out image
)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(out[:,:,::-1])
```
