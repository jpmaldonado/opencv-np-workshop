# What is features and why do we need them?

- "Important" parts of an image:
    - Importance means *how useful a feature is to describe what is on a scene*

Features consist of two parts:

- **Keypoints:** Special points of an image, where special means edges/color changes, corners.
- **Descriptors:** Describe (as a vector) what is happening on a given keypoint, unambiguously.

We look for changes in the intensity function $I(x,y)$.

```python
img = cv2.imread("../opencv-np-workshop/data/img/regensburg.jpg")
```

## Harris corner detection

```python
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
corners = cv2.cornerHarris(img_gray, blockSize=2, ksize=3, k=0.04)

# blockSize = Neighbourhood size
# ksize = window for Sobel operator (usually 3)
# k = characteristic for the linear approximation. Usually small, 0.1 or less
```

```python
corners = cv2.dilate(corners, None)
```

```python
img2 = img.copy()
```

```python
img2[corners>0.1*corners.max()] = [0,0,255] # Red to some fraction of the corners
```

```python
#plt.figure(figsize=(20,20))
plt.imshow(img2[:,:,::-1])
```
