```python
import cv2
```

```python
cam = cv2.VideoCapture("./data/video/traffic.mp4") # camera id
```

```python
while True:
    ret, frame = cam.read() #ret = T/F flag
    
    if ret:
        # Get a frame, do something with it
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        cv2.imshow("Gray traffic", gray)
    
    # Add some way-out of infinite loop
    if cv2.waitKey(1) == 27: # 27 == ESC
        break
        
cam.release() # close connection with cam
cv2.destroyAllWindows()
```

# Save video

- Define codec for output video
- Create a target for the output video
- Read frame from original ==> Transform ==> Store frame (frame by frame)
- **Same workflow for camera/video file/GIF files**

```python
cam = cv2.VideoCapture("./data/video/traffic.mp4")
fourcc = cv2.VideoWriter_fourcc(*'DIVX') # DIVX usually good in Windows
out = cv2.VideoWriter("./data/video/traffic-gray.avi"
                        , fourcc
                         , 25
                         , (1920,1080)
                         , isColor = False)

while cam.isOpened():
    ret, frame = cam.read() #ret = T/F flag
    if ret:
        # Get a frame, do something with it
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        out.write(gray)
    else:
        break
cam.release() # close connection with cam
out.release() # Release connection with output file
```
