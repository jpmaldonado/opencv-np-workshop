```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("../opencv-np-workshop/data/img/lane.jpg", 0)
plt.imshow(img, cmap="gray")
```

**Goal:** detect edges in that image.


- Edges $\approx$ derivatives

## Calculating first derivatives

```python
dx = cv2.Sobel(img, -1, 1, 0, ksize=3)
dy = cv2.Sobel(img, -1, 0, 1, ksize=3)
# First argument: -1 = ddepth = type of array returned, if -1 keeps type of original array
# Second argument =  derivative on x
# Third argument =  derivative on y
```

```python
plt.imshow(dx, cmap="gray")
```

```python
plt.imshow(dy, cmap="gray")
```

```python
laplacian = cv2.Laplacian(img, -1)
plt.imshow(laplacian, cmap="gray")
```

## Canny edge detection

```python
edges = cv2.Canny(img, 200, 250, 3) #
plt.imshow(edges, cmap="gray")
```
