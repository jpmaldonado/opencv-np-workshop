```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

# Use a pre-trained model

```python
img = cv2.imread("../opencv-np-workshop/data/img/travolta.jpg")
```

```python
plt.imshow(img[:,:,::-1])
```

```python
face_cascade_file = "../opencv-np-workshop/data/haarcascades/haarcascade_frontalface_alt.xml"
```

```python
face_cascade = cv2.CascadeClassifier()
```

```python
face_cascade.load(face_cascade_file)
```

```python
rect = face_cascade.detectMultiScale(img)
```

```python
rect # Coordinates of the bounding rectangle
```

```python
x, y, w, h = rect[0]
plt.imshow(cv2.rectangle(img, (x,y),(x+w,y+h), (0,255,0), 5)[:,:,::-1])
```
