```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
nemo = cv2.imread("../opencv-np-workshop/data/img/nemo.jpg")
```

```python
plt.imshow(nemo[:,:,::-1])
```

```python
nemo_hsv = cv2.cvtColor(nemo, cv2.COLOR_BGR2HSV)
```

**Idea:** Define a range of oranges in HSV, create a mask and use that mask to find Nemo.

```python
import numpy as np
```

```python
lower = np.array([10, 100, 100])
upper = np.array([30, 255, 255])
```

```python
mask = cv2.inRange(nemo_hsv, lower, upper) # Create mask with a correct range of values
```

```python
res = cv2.bitwise_and(nemo, nemo, mask=mask) # Add the frame with itself using that mask
```

```python
plt.imshow(res[:,:,::-1])
```
