```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("../opencv-np-workshop/data/img/art/100532.jpg") #100532
```

```python
img.shape
```

```python
plt.imshow(img[:,:,::-1])
```

```python
hist = cv2.calcHist([img], [0], None, [255],[0,255])
# img, channel, mask, n_bins, [min_val, max_val]
# Channel: 0 = blue, 1=green, 2=red
```

```python
plt.plot(hist)
```

### Look at distribution of all channels on the same plot

```python
colors = ('b','g','r')
for i, col in enumerate(colors):
    hist = cv2.calcHist([img],[i], None, [255],[0,255])
    plt.plot(hist, color=col)
    plt.xlim([0,255])
plt.show()
```

## Histogram equalization

```python
gray = cv2.imread("../opencv-np-workshop/data/img/wiki.jpg", 0) #Read as grayscale
```

```python
hist = cv2.calcHist([gray],[0], None, [255],[0,255])
plt.plot(hist)
```

```python
equ = cv2.equalizeHist(gray)
```

```python
plt.imshow(equ, cmap="gray")
```

```python
eq_hist = cv2.calcHist([equ],[0],None,[255],[0,255])
plt.plot(eq_hist)
```
